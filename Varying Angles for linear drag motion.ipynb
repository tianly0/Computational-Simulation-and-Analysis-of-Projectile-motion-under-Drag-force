```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#parameters
u = 40 # Initial Velocty 
theta_deg = 60 # Angle of Lauching 
g = 9.8
```

```python
angles = [20, 30, 35, 45, 60, 75]

k = 0.03
dt = 0.01

for angle in angles:

    theta = np.radians(angle)

    vx = u * np.cos(theta)
    vy = u * np.sin(theta)
    v = np.sqrt(vx**2 + vy**2)

    x = 0
    y = 0

    X = []
    Y = []

    while y >= 0:

        ax = -k * v * vx
        ay = -g - (k * v * vy)

        vx += ax * dt
        vy += ay * dt

        x += vx * dt
        y += vy * dt

        X.append(x)
        Y.append(y)

    plt.plot(X, Y, label=f"{angle}°")

plt.xlabel("Horizontal Distance")
plt.ylabel("Vertical Distance")
plt.title("Projectile Motion with Linear Drag (Multiple Angles)")
plt.legend()
plt.grid(True)
plt.show()
```
