```python
k = [0.05, 0.04, 0.03, 0.02, 0.01]
angle = 30 
theta = np.radians(angle)

dt = 0.01

for k in k:

    vx = u * np.cos(theta)
    vy = u * np.sin(theta)
    v= np.sqrt(vx**2+ vy**2)

    x = 0
    y = 0

    X = []
    Y = []

    while y >= 0:

        ax = -k * v *vx
        ay = -g - (k * v *vy)

        vx += ax * dt
        vy += ay * dt

        x += vx * dt
        y += vy * dt

        X.append(x)
        Y.append(y)

    plt.plot(X, Y, label=f"k={k}")

plt.xlabel("Horizontal Distance")
plt.ylabel("Vertical Distance")
plt.title("Projectile Motion with Linear Drag (Varying values of K)")
plt.legend()
plt.grid(True)
plt.show()
```
