```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m =1 #kg
u = 40 # m/s
g = 9.8 # m/s^2
theta_degree = 30
theta = np.radians(theta_degree)
```

```python
T = (2*u*np.sin(theta))/g
```

```python
t = np.linspace(0, T, 500)
```

```python
vx = u * np.cos(theta)
vy = u * np.sin(theta) - g*t
v = np.sqrt(vx**2 + vy**2)
```

```python
KE1 = 0.5* m *v**2 * np.ones_like(t)
```

```python
# Plot of KE vs Time for No drag


plt.figure(figsize = (8,5))

plt.plot(t, KE1)
plt.xlabel("Time(sec)")
plt.ylabel("Kinetic Energy(J)")
plt.title("KE vs t (For no drag)")


plt.grid(True)
plt.show()
```

```python
vx = u * np.cos(theta)
vy = u * np.sin(theta)
```

```python
x = 0
y = 0
t1 = 0
```

```python
times1 = []
KE2 = []
k1 = 0.05
dt = 0.01
```

```python
while y>=0:
    v = np.sqrt(vx**2 + vy**2)
    times1.append(t1)
    KE2.append(0.5*m*v**2)

    ax = -k1*vx
    ay = -g - k1*vy

    vx += ax*dt
    vy += ay*dt

    x += vx*dt
    y += vy*dt
    t1 += dt
```

```python
# Plot of KE vs Time for Linear drag

plt.figure(figsize=(8,5))
plt.plot(times1, KE2)

plt.xlabel("Time (sec)")
plt.ylabel("Kinetic Energy (Joules)")
plt.title("KE vs Time(Linear Drag)")
plt.grid(True)
plt.show()
```

```python
vx = u * np.cos(theta)
vy = u * np.sin(theta)
```

```python
x = 0
y = 0
t2 = 0
```

```python
times2 = []
KE3 = []
k2 = 0.05
dt = 0.01
```

```python
while y>=0:
    v = np.sqrt(vx**2 + vy**2)
    times2.append(t2)
    KE3.append(0.5*m*v**2)

    ax = -k2*v*vx
    ay = -g - k2*v*vy

    vx += ax*dt
    vy += ay*dt

    x += vx*dt
    y += vy*dt
    t2 += dt
```

```python
# Plot of KE vs Time for quadratic drag


plt.figure(figsize=(8,5))
plt.plot(times2, KE3)


plt.xlabel("Time (sec)")
plt.ylabel("Kinetic Energy (Joules)")

plt.title("KE vs Time (Quadratic Drag)")
plt.grid(True)
plt.show()
```

```python
# Plot of KE vs Time for all cases

k=0
k1=k1
k2=k2
plt.figure(figsize = (8,5))

plt.plot(t, KE1,label=f"No drag (k={k})")
plt.plot(times1, KE2, label=f"Linear drag (k={k1})")
plt.plot(times2, KE3, label=f"Quadratic Drag (k={k2})")

plt.xlabel("Time (sec)")
plt.ylabel("Kinetic Energy (Joules)")
plt.title("KE vs time")

plt.legend()
plt.grid(True)
plt.show()
```
