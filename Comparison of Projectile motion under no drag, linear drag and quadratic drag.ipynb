```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Variables

u = 40 # initial velocity
g = 9.8
theta_degree = 30 # launch angle
theta = np.radians(theta_degree)
```

```python
#Time of flight
T = (2*u*np.sin(theta))/g
```

```python
#For no drag force

t = np.linspace(0, T, 500)

X1 = u * np.cos(theta) * t
Y1 = (u * np.sin(theta) * t) - (0.5 * g * t**2)
```

```python
#For linear drag force

k1 = 0.5
dt = 0.01

vx = u * np.cos(theta)
vy = u * np.sin(theta)

x = 0
y = 0

X2 = [ ]
Y2 = [ ]
t1=0
time1 =[]
count1=0
while y>=0:
    
    time1.append(t1)
    ax = - k1* vx
    ay = -g - k1*vy

    vx += ax * dt
    vy += ay * dt

    x += vx * dt
    y += vy * dt

    t1 +=dt

    X2.append(x)
    Y2.append(y)
    count1 += 1

```

```python
#For Quadratic Drag

k2 = 0.1
dt = 0.01

vx = u * np.cos(theta)
vy = u * np.sin(theta)
v = np.sqrt(vx**2 + vy**2)

x = 0
y = 0

X3 = [ ]
Y3 = [ ]
t2 = 0
time2 =[]
count2=0
while y>=0:

    time2.append(t2)
   
    ax = -k2 * v * vx
    ay = -g -(k2 * v * vy)

    vx += ax * dt
    vy += ay * dt

    x += vx * dt
    y += vy * dt

    t2 +=dt    
    X3.append(x)
    Y3.append(y)
    
    count2 += 1

    
```

```python
plt.figure(figsize=(10,5))
k1 = k1
k2 = k2


plt.plot(X1, Y1, label=f"No Drag, k=0")
plt.plot(X2, Y2, label=f"Linear Drag, k= {k1}")
plt.plot(X3, Y3, label=f"Quadratic Drag, k= {k2}")

plt.xlabel("Horizontal Distance (m)")
plt.ylabel("Vertical Distance (m)")

plt.title("Projectile Motion Comparison")

plt.legend()
plt.grid(True)

plt.show()
```

```python
plt.figure(figsize = (8,5))

plt.plot(t, X1, label=f"No drag")
plt.plot(time1, X2,label=f"Linear drag")
plt.plot(time2, X3,label=f"Quadratic drag")

plt.ylabel("Horizontal Distance(m)")
plt.xlabel("Time(sec)")
plt.grid(True)
plt.legend()

plt.show()
```

```python
print(f"Time taken for no drag = {T:.2f}")
print(f"Time taken for linear drag = {time1[count1-1]:.2f}")
print(f"Time taken for quadratic drag = {time2[count2-1]:.2f}")
```

```python
plt.figure(figsize = (8,5))

plt.plot(t, Y1, label=f"No drag")
plt.plot(time1, Y2,label=f"Linear drag")
plt.plot(time2, Y3,label=f"Quadratic drag")

plt.ylabel("Vertical Distance(m)")
plt.xlabel("Time(sec)")
plt.grid(True)
plt.legend()

plt.show()
```
