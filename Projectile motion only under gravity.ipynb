```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#parameters
u = 40 # Initial Velocty 
theta_deg = 60 # Angle of Lauching 
g = 9.8
```

```python
theta = np.radians(theta_deg)
```

```python
T = (2*u* np.sin(theta))/g # Time of flight
```

```python
t = np.linspace(0, T, 2000) # time
```

```python
x = (u* np.cos(theta)) *t # Horizontal Position
y = (u* np.sin(theta) *t) -(0.5* g*  t**2) # Vertial Position 
```

```python
Vx = u * np.cos(theta) * np.ones_like(t)  # Horizontal Velocity 
Vy = (u * np.sin(theta) ) - (g * t) # Vertical Velocity
```

```python
H = (u**2 * np.sin(theta)**2)/(2 * g) # Maximum Height
```

```python
R = (u**2 * np.sin(2*theta))/g # Range 
```

```python
print(f"Initial Velocity = {u} m/s")
print(f"Angle of Lauching  = {theta_deg} degree")
print(f"Time of flight = {T:.2f} seconds")
print(f"Maximum Height = {H:.2f} meters")
print(f"Range = {R:.2f} meters")
```

```python
angle = 60
velocity = 50
plt.figure(figsize=(8,5))

plt.plot(x, y, label=f"{angle}°, {velocity}m/s")

plt.xlabel("Horizontal Distance")
plt.ylabel("Vertical Distance")

plt.title("Projectile Trajcectory")
plt.legend()
plt.grid(True)

plt.show()
```
